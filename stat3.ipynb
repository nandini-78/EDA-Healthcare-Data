1. What is the average age of patients who have had a stroke and also have hypertension?
(Round up the final answer to 2 decimal places.)  

```python
round(df[(df['stroke']==1)&(df['hypertension']==1)]['age'].mean(),2)
```

2. What percentage of patients who have ever been married and have a heart disease have had a stroke?
 (Round up the final answer to 2 decimal places.)  

```python
round((df[(df['stroke']==1)&(df['heart_disease']==1)&(df['ever_married']=='Yes')].shape[0]/df.shape[0])*100,2)
```

3. How many patients have a BMI greater than 30 and have never smoked? 
(Integer Format)

```python
df[(df['smoking_status']=='never smoked')&(df['bmi']>30)]['smoking_status'].count()
```

4.  What is the correlation between average glucose level and age for patients who have had a stroke?
 (Round up the final answer to 2 decimal places.)

```python
round(df[df['stroke']==1][['avg_glucose_level','age']].corr().iloc[0,1],2)
```

5.  How many patients have a BMI greater than 25 and have hypertension, but have never been married?

6.  What percentage of patients who have had a stroke and have a heart disease are male? 
(Round up the final answer to 2 decimal places.)

```python
round((df[(df['stroke']==1)&(df['heart_disease']==1)&(df['gender']=='Male')].shape[0]/df.shape[0])*100,2)
```

7. Check the normality of the BMI column by using the Shapiro-Wilk test. If it is a normal distribution provide an answer as “Normal Distribution” else provide the answer as ”Not A Normal Distribution”.  

```python
df['bmi'].isnull().sum()
```

```python
df['bmi'].replace(df['bmi'].mean(),inplace=True)
```

```python
from os import P_ALL
import scipy.stats as st
s,p_=st.shapiro(df['bmi'])
s,p_
```

```python
p=0.05
if p_>0.05:
  print("Normal Distribution")
else:
  print("Not A Normal Distribution")
```
